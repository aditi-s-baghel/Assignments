# Q1. Create a Player Class

## Problem Statement
Create a class named `Player` using constructor.

Store:
- Player Name
- Jersey Number

Create a method to display player details.

Create two objects.

## Sample Output
```python
Player Name: Virat
Jersey Number: 18

Player Name: Rohit
Jersey Number: 45
```


````python
class Player:
    def __init__(self,name,num):
        self.name = name
        self.num = num
        
    def display(self):
        print("PLayer Name: ",self.name)
        print("Jersey Number",self.num)

obj1 = Player("Virat",18)
obj2 = Player("Rohit",45)
obj1.display()
obj2.display()
````

# Q4. Circle Class

## Problem Statement
Create a class named `Circle`.

Use constructor to take radius.

Create methods to calculate:
- Area
- Circumference

Use:
```python
3.14 * r * r
2 * 3.14 * r
```

## Sample Output
```python
Area: 78.5
Circumference: 31.400000000000002
```


````python
class Circle:
    def __init__(self,radius):
        self.radius = radius

    def calculator(self):
        Area = 3.14 * self.radius * self.radius
        Circumference = 2 * 3.14 * self.radius
        print("Area:", Area)
        print("Circumference:", Circumference)

obj = Circle(5)
obj.calculator()
````

# Q7. Bank Account Class

## Problem Statement
Create a class named `BankAccount`.

Store:
- Account Holder Name
- Balance

Create method `deposit()` to add money into balance.

## Sample Output
```python
Name: Arjun
Balance After Deposit: 55000
```


````python
class BankAccount:
    def __init__(self,name,bal):
        self.name = name
        self.bal = bal
    
    def deposit(self,money):
        self.money = money
        Money_Deposited= self.money + self.bal
        print("Name:",self.name)
        print("Balance After Deposit:",Money_Deposited)

obj = BankAccount("Arjun",0)
obj.deposit(55000)
````

# Q10. Food Item Class

## Problem Statement
Create a class named `FoodItem`.

Store:
- Food Name
- Price
- Quantity

Create method to calculate total bill.

Formula:
```python
Total = Price * Quantity
```

## Sample Output
```python
Food Name: Pizza
Total Bill: 800
```


````python
class FoodItem:
    def __init__(self,name,price,quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def calculator(self):
        Total = self.price * self.quantity

        print("Food Name:",self.name)
        print("Total Bill:",Total)

obj = FoodItem("Pizza",400,2)
obj.calculator()
````
